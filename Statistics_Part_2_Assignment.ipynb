Que.2 What is the null hypothesis, and how does it differ from the alternative
      hypothesis?

  ->  In statistical hypothesis testing, the null hypothesis (H0) proposes  
      that there is no significant effect or relationship between variables, while the alternative hypothesis (H1 or Ha) suggests that there is a significant effect or relationship. Essentially, the null hypothesis is what you're trying to disprove, and the alternative hypothesis is what you're trying to prove.

      Null Hypothesis (H0):
      1.The default assumption that there is no effect or difference.
      2.It's the statement of no change, no relationship, or no
        difference between groups.
      3.It's the hypothesis that we are trying to disprove.
      4.It assumes any observed effect is due to random chance.

      Alternative Hypothesis (H1 or Ha):
      1.The statement that contradicts the null hypothesis.
      2.It suggests that there is a significant effect or difference.
      3.It's the hypothesis we are trying to prove and which is accepted if
        we have enough evidence to reject the null hypothesis.
      4.It assumes the observed effect is due to a non-random cause.

      Example:
      If we are testing a new fertilizer.

      Null Hypothesis (H0):
      The new fertilizer has no effect on plant growth compared to
      the standard fertilizer.

      Alternative Hypothesis (H1):
      The new fertilizer does have an effect on plant growth
      (either increases or decreases it) compared to the standard fertilizer.

      In essence: The null hypothesis is the "no effect" or
      "no difference" baseline, and the alternative hypothesis is the claim that there is a real effect or difference.

      How they relate:
      1.The null and alternative hypotheses are mutually exclusive.
      2.If the null hypothesis is rejected, it means there's sufficient
        evidence to support the alternative hypothesis.
      3.You can't prove the null hypothesis, you can only fail to reject it.
      4.The alternative hypothesis is often what the researcher is actually
        trying to demonstrate.




Que.3 What is the significance level in hypothesis testing, and why is it
      important?

  ->  In hypothesis testing, the significance level (often denoted by α) is  
      the probability of incorrectly rejecting the null hypothesis when it is actually true (a Type I error). It sets the threshold for determining statistical significance. It's important because it helps researchers decide whether the observed data provides enough evidence to reject the null hypothesis and conclude that the alternative hypothesis is more likely.

      Defining Significance:
      A significance level, or alpha (α), is the probability of
      concluding that a difference or effect exists when, in reality, it
      does not. It's essentially the risk of a false positive, a Type I
      error. For instance, if α = 0.05 (5%), it means there's a 5% chance
      of incorrectly rejecting the null hypothesis when it is actually true.

      Choosing the Significance Level:
      Researchers choose the significance level before conducting the
      test, typically based on factors like the severity of the
      potential consequences of a Type I error, the sample size, and the
      field of study. Common values include 0.05 (5%), 0.01 (1%), or 0.10 (10%).

      Relationship to P-value:
      The p-value is calculated based on the observed data and is compared
      to the significance level to make a decision about the null
      hypothesis. If the p-value is less than or equal to the
      significance level, the null hypothesis is rejected.

      Why it's important:
      The significance level is crucial because it helps researchers
      make informed decisions about the data and draw valid conclusions.
      By setting a threshold for statistical significance, they can
      control the risk of making a Type I error and avoid
      incorrectly concluding that an effect exists when it doesn't.

Que.4 What does a P-value represent in hypothesis testing?

  ->  In hypothesis testing, a p-value represents the probability of observing
      data as extreme as, or more extreme than, the data actually observed, assuming the null hypothesis is true. It's a measure of the strength of evidence against the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis.

      Null Hypothesis:
      This is the assumption that there is no effect or difference in
      the population being studied.

      P-value Calculation:
      The p-value is calculated by determining the probability of
      observing the test statistic (the result of a statistical test) or
      a more extreme value, assuming the null hypothesis is true.

      Significance Level (Alpha):
      This is a predetermined probability threshold (often 0.05 or 5%) used
      to decide whether to reject the null hypothesis.

      Interpreting the P-value:

      If p-value ≤ significance level: The result is considered
      statistically significant, and the null hypothesis is rejected.
      This suggests that the observed effect is unlikely to have occurred by chance.

      If p-value > significance level: The result is not statistically significant, and the null hypothesis is not rejected. This means
      that the observed effect could be due to random chance.

      In simpler terms: Imagine you're flipping a coin and want to test
      if it's fair. The null hypothesis is that the coin is fair. If you
      flip it 10 times and get 9 heads, the p-value represents the
      probability of getting that many heads or more if the coin were fair.
      A low p-value (e.g., <0.05) would make you doubt the coin is fair
      and lead you to reject the null hypothesis, suggesting it's biased.

Que.6 What are Type 1 and Type 2 errors in hypothesis testing?

   -> In hypothesis testing, a Type 1 error (false positive) occurs when you
      incorrectly reject a true null hypothesis, while a Type 2 error (false negative) happens when you fail to reject a false null hypothesis. Essentially, a Type 1 error is saying something is happening when it's not, and a Type 2 error is saying something isn't happening when it actually is.

      Type 1 Error (False Positive):
      This error occurs when you conclude that there's a significant effect
      or difference in your data when, in reality, there isn't one.
      In other words, you reject the null hypothesis (which states there's
      no effect) when it's actually true.

      Type 2 Error (False Negative):
      This error occurs when you fail to detect a real effect or difference
      in your data, even though one exists. You're essentially failing
      to reject the null hypothesis when it's actually false.

      Example:
      Imagine testing a new drug. A Type 1 error would be claiming the drug
      is effective when it actually isn't, while a Type 2 error would
      be concluding the drug isn't effective when it actually is.

Que.7 What is the difference between a one-tailed and a two-tailed test in
      hypothesis testing?

  ->  In hypothesis testing, a one-tailed test checks for a difference in one
      specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (e.g., greater than or less than, or simply different). A two-tailed test is more conservative, requiring a stronger effect to be detected as significant than a one-tailed test.

     

      One-tailed test:

      Directional:
      The alternative hypothesis specifies a particular direction of
      the difference (e.g., "the new drug is more effective than the old drug").

      Example:
      If you hypothesize that a new drug will increase blood pressure,
      you would use a one-tailed test.

      Critical Region:
      The critical region (the area where the null hypothesis is rejected)
      is located in one tail of the distribution.

      Power:
      Can be more powerful than a two-tailed test if the direction of
      the effect is correctly specified.

      Two-tailed test:

      Non-directional:
      The alternative hypothesis simply states that the parameter is
      different from the null hypothesis, without specifying the direction
      of the difference (e.g., "the new drug has a different effect on
      blood pressure than the old drug").

      Example:
      If you hypothesize that a new drug has a different effect on
      blood pressure than the old drug, you would use a two-tailed test.

      Critical Region:
      The critical region is split between both tails of the distribution.

      Power:
      Less powerful than a one-tailed test when the direction of the effect
      is known, but more conservative and less likely to make a false positive.

      When to use which:
      One-tailed:
      Use when you have a strong prior belief about the direction of
      the effect and you're confident that the effect will only go in
      that direction.

      Two-tailed:
      Use when you're unsure about the direction of the effect, or when
      you're primarily interested in detecting any difference, regardless
      of direction.


Que.8 What is the Z-test, and when is it used in hypothesis testing?

   -> A Z-test is a statistical test used in hypothesis testing to determine  
      if there's a significant difference between two sample means or if a sample mean is significantly different from a known population mean. It's used when the population standard deviation is known and the sample size is large (typically n ≥ 30), allowing the use of the normal distribution to approximate the test statistic, according to DataCamp.

      When to use a Z-test:

      Comparing a sample mean to a known population mean:
      You want to see if the average of your sample is different from
      the known average of the entire population.

      Comparing two sample means:
      You want to determine if there's a significant difference between
      the averages of two different groups.

      Large sample sizes and known population standard deviation:
      The Z-test is most appropriate when the sample size is large
      (n ≥ 30) and the population standard deviation is known.

      Key conditions for using a Z-test:

      Normally distributed population (or large sample size):
      The data should be approximately normally distributed, or if the
      sample size is large enough, the Central Limit Theorem can be used
      to approximate the distribution of the sample means as normal.

      Known population standard deviation:
      You need to know the standard deviation of the population from which
      the sample was drawn.

      Hypothesis testing with Z-tests:

      The Z-test follows the general steps of hypothesis testing:
      1. State the null hypothesis (H0):
         This is a statement that there is no significant difference
         between the sample mean and the population mean
         (or between the two sample means).

      2. State the alternative hypothesis (H1):
         This is a statement that there is a significant difference.

      3. Set the significance level (α):
         This is the probability of rejecting the null hypothesis when
         it is actually true (usually 0.05 or 0.01).
      4. Calculate the test statistic (Z-score):
         This measures how many standard deviations the sample mean is
         away from the hypothesized population mean.

      5. Determine the p-value:
         This is the probability of obtaining a test statistic as extreme
         as, or more extreme than, the one observed, assuming the
         null hypothesis is true.

      6. Make a decision:
         If the p-value is less than or equal to the significance level
         (α), you reject the null hypothesis and conclude that there is
         a statistically significant difference.


Que.9 How do you calculate the Z-score, and what does it represent in
      hypothesis testing?

  ->  The z-score, or z-statistic, is a numerical value that indicates how  
      many standard deviations a specific data point is away from the mean of a population. It's calculated using the formula: Z = (X - μ) / σ, where X is the raw score, μ is the population mean, and σ is the population standard deviation. In hypothesis testing, the z-score helps determine if a sample mean is significantly different from a hypothesized population mean, aiding in making inferences about the population based on the sample data.

      Calculation of the Z-score:

      1.Identify the raw score (X): This is the specific data point you want  
        to compare to the population mean.
      2.Find the population mean (μ): This is the average of all data points  
        in the population.
      3.Find the population standard deviation (σ): This measures the spread  
        of data points around the population mean.
      4.Apply the formula: Subtract the population mean from the raw score,
        then divide the result by the population standard deviation.

      Representation in Hypothesis Testing:

      Z-test:
      The z-score is used in z-tests, a type of statistical hypothesis
      test, to determine if a sample mean is significantly different
      from a hypothesized population mean.

      Comparing sample mean to population mean:

      The z-score indicates how many standard deviations the sample
      mean is away from the population mean.

      Determining significance:
      By comparing the calculated z-score to a critical value
      (determined by the chosen significance level), researchers can
      decide whether to reject the null hypothesis (which typically
      states there's no significant difference between the sample mean
      and population mean).

      Example:
      If a consumer group tests 100 phones and finds an average battery
      life of 11.8 hours, with a known population standard deviation of
      0.5 hours, and the company claims the average battery life is
      12 hours, a z-test can be performed. The z-score would be
      calculated, and compared to a critical value to see if there's
      enough evidence to reject the company's claim.

       Small Sample Sizes:

       more appropriate in these cases because it accounts for the
       increased variability in the sample standard deviation.

       Unknown Population Standard Deviation:
       When the population standard deviation (σ) is unknown, you'll
       need to estimate it from the sample data. This introduces
       uncertainty, and the T-distribution is designed to handle this uncertainty.

       When to use the Normal Distribution (Z-distribution):

       Large Sample Sizes:
       When the sample size is 30 or more, the T-distribution and
       the Z-distribution become very similar, and the Z-distribution can
       be used with reasonable accuracy.

       Known Population Standard Deviation:
       If the population standard deviation is known, you can use
       the Z-distribution because there's no need to estimate it from the sample.

Que.14 What is a confidence interval, and how is it used to interpret
       statistical results?
    


   ->  A confidence interval is a range of values used to estimate an unknown
       population parameter based on sample data. It expresses the probability that a parameter will fall within a specific range, providing a degree of uncertainty or certainty in a statistical estimate. Confidence intervals are used to interpret statistical results by helping determine the significance of findings and the precision of estimates.

   
       What is a Confidence Interval?

       Range of Values:
       Instead of presenting a single point estimate
       (like the average height of students), confidence intervals provide
       a range of values where the true population value is likely to fall.

       Based on Sample Data:
       These intervals are calculated from sample data, meaning they are
       not exact representations of the true population parameter,
       but rather estimates based on a subset of the population.

       Confidence Level:
       The confidence level (e.g., 95%, 99%) represents the percentage
       of times that the interval would contain the true population
       parameter if the study were repeated many times.

       Uncertainty:
       Confidence intervals help quantify the uncertainty around
       sample estimates, giving an idea of how precise the estimate is.

       How to Interpret Statistical Results with Confidence Intervals:

       1. Significance Testing:
          A 95% confidence interval can be used to assess
          statistical significance. If the interval does not contain the
          null hypothesis value (often zero), the finding is
          considered statistically significant. For example, if a
          confidence interval for the difference in two group means does
          not include zero, it suggests a statistically significant
          difference between the groups.

       2. Precision of Estimates:
          The width of the confidence interval reflects the precision of
          the estimate. A narrow interval indicates a more precise
          estimate, while a wider interval suggests greater uncertainty.

       3. Effect Size:
          Confidence intervals can provide insights into the magnitude of
          an effect. A wider interval might suggest a smaller effect or a
          more uncertain estimate of the effect.

       4. Long-term Frequency:
          The confidence level indicates the long-term frequency with
          which the interval would contain the true population
          parameter across repeated sampling. For example, a 95%
          confidence interval means that if you were to repeat the study
          many times, you would expect the true population parameter to
          fall within the interval 95% of the time.

       5. Not a Probability of the Interval Containing the True Value:
          It's crucial to understand that a 95% confidence interval does
          not mean there's a 95% probability that the current
          interval contains the true parameter. It's about the
          long-term frequency of capturing the true parameter over many samples.

The assumptions of ANOVA include independent random sampling, normality of data within each group, and homogeneity of variance (equal variance across groups).

       What is ANOVA?
       ANOVA is a parametric test that helps determine if
       there's a statistically significant difference between the means
       of three or more groups.
       It tests the null hypothesis that all group means are equal,
       against the alternative hypothesis that at least one group mean
       is different.
       ANOVA works by comparing the variance within each group to the
       variance between the group means.

       Assumptions of ANOVA:
       Independent Random Sampling: Observations within each group and
       between groups must be independent. This means that one
       observation doesn't influence another.

       Normality: The data within each group should be approximately
       normally distributed.

       Homogeneity of Variance: The variance (spread) of the data should
       be approximately equal across all groups. This is also known
       as homoscedasticity.

       Why are these assumptions important?
       These assumptions ensure that the ANOVA test is valid and that
       the results are reliable.
       If the assumptions are violated, the results of the ANOVA test
       may be inaccurate or misleading.

       Types of ANOVA:
       One-way ANOVA: Used when there's one independent variable with
       multiple levels (groups).

       Two-way ANOVA: Used when there are two or more independent variables.

       Factorial ANOVA: A type of ANOVA that examines the effects of
       multiple independent variables on a dependent variable.

Que.22 What is the F-test, and how does it relate to hypothesis testing?

  ->   The F-test is a statistical test used in hypothesis testing to  
       determine if the variances of two populations are equal or if a regression model is significant. It calculates an F-statistic by comparing the variance between groups to the variance within groups, and then uses the F-distribution to determine if the calculated F-statistic is statistically significant.


       Purpose:
       The F-test is particularly useful in comparing variances between two
       or more groups, often used in analysis of variance (ANOVA).
       It can also assess the overall significance of a regression model
       by testing if the model's parameters are significantly different from zero.

       F-statistic:
       The F-statistic is the ratio of two variances: the variance between
       the groups being compared (also known as the "explained variance"
       in regression) to the variance within each group
       (the "unexplained variance").

       Hypothesis Testing:
       Null Hypothesis: In ANOVA, the null hypothesis typically states
       that the means (and hence, the variances) of the groups are equal.
       In regression, the null hypothesis usually states that all the
       model parameters are zero.

       Alternative Hypothesis: The alternative hypothesis states that
       the means (or model parameters) are not equal, or that at least
       one parameter is not zero.

       F-distribution:
       The calculated F-statistic is compared to critical values from
       the F-distribution to determine its significance. The F-distribution
       is a probability distribution that is used to assess the likelihood
       of observing the calculated F-statistic if the null hypothesis is true.

       Decision Rule:
       If the calculated F-statistic is greater than the critical value
       (or if the p-value is less than the significance level), the
       null hypothesis is rejected, suggesting that there is a
       significant difference between the variances or that the model
       is statistically significant.

```python
#Que.3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (must be between 0 and 1)

    Returns:
    float: Probability P(X = x)
    """
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0.0  # Invalid value for Bernoulli

# Example usage
p = 0.7
print("P(X = 0):", bernoulli_pdf(0, p))
print("P(X = 1):", bernoulli_pdf(1, p))
print("P(X = 2):", bernoulli_pdf(2, p))  # Invalid outcome
```

```python
#Que.10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(sample_size, num_samples, distribution, **params):
    """
    Simulate the Central Limit Theorem by sampling from a non-normal distribution and plotting sample means.

    Parameters:
    sample_size (int): The size of each sample.
    num_samples (int): The number of samples to take.
    distribution (str): The type of non-normal distribution ('exponential' here).
    **params: Parameters to pass to the distribution (e.g., scale for exponential).

    Returns:
    np.ndarray: Array of sample means.
    """
    sample_means = []

    for _ in range(num_samples):
        if distribution == 'exponential':
            sample = np.random.exponential(**params, size=sample_size)
        elif distribution == 'uniform':
            sample = np.random.uniform(**params, size=sample_size)
        else:
            raise ValueError("Unsupported distribution type")

        sample_means.append(np.mean(sample))

    return np.array(sample_means)

def plot_distribution_of_means(sample_means):
    """
    Plot the distribution of sample means to visualize the Central Limit Theorem.

    Parameters:
    sample_means (np.ndarray): Array of sample means.
    """
    plt.figure(figsize=(8, 5))
    plt.hist(sample_means, bins=50, density=True, edgecolor='black', alpha=0.7)
    plt.title('Distribution of Sample Means (CLT Demonstration)')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

# Parameters
sample_size = 30  # Each sample contains 30 data points
num_samples = 1000  # Take 1000 samples
distribution_type = 'exponential'  # Non-normal distribution
params = {'scale': 1}  # Exponential distribution scale parameter

# Simulate CLT
sample_means = central_limit_theorem(sample_size, num_samples, distribution_type, **params)

# Plot the distribution of sample means
plot_distribution_of_means(sample_means)
```

```python
#Que.15 Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for hypothesis testing.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The mean of the population under the null hypothesis.
    population_std (float): The standard deviation of the population.
    alpha (float): The significance level (default is 0.05).

    Returns:
    float: Z-statistic value
    float: p-value
    bool: Result of hypothesis test (True if null hypothesis is rejected)
    """
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

    # Check if we reject the null hypothesis
    reject_null = p_value < alpha

    return z_statistic, p_value, reject_null

def plot_z_test_distribution(sample_data, population_mean, z_statistic):
    """
    Plot the standard normal distribution with the Z-statistic.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The population mean under the null hypothesis.
    z_statistic (float): The calculated Z-statistic value.
    """
    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Standard Normal Distribution")

    # Shade the region of rejection
    plt.fill_between(x, y, where=(x >= z_statistic), color='red', alpha=0.5, label=f'Rejection Region (Z > {z_statistic:.2f})')

    # Highlight the Z-statistic on the plot
    plt.axvline(z_statistic, color='blue', linestyle='--', label=f'Z-statistic = {z_statistic:.2f}')

    plt.title('Z-Test: Hypothesis Testing')
    plt.xlabel('Z')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage:

# Sample data (e.g., test scores)
sample_data = np.array([85, 88, 91, 87, 85, 92, 89, 84, 90, 91])

# Population parameters under the null hypothesis
population_mean = 90  # Population mean (null hypothesis)
population_std = 5   # Population standard deviation

# Perform Z-test
z_statistic, p_value, reject_null = z_test(sample_data, population_mean, population_std)

# Output the results
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print("Reject the null hypothesis?" , "Yes" if reject_null else "No")

# Plot the distribution
plot_z_test_distribution(sample_data, population_mean, z_statistic)
```

```python
#Que.16 Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    sample_data (array-like): The sample data.
    confidence_level (float): The desired confidence level (default is 95%).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation
    sample_size = len(sample_data)

    # Z-value for the given confidence level (for normal distribution)
    z_value = stats.norm.ppf((1 + confidence_level) / 2)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Confidence interval calculation
    margin_of_error = z_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset
sample_data = np.array([85, 88, 91, 87, 85, 92, 89, 84, 90, 91])

# Calculate the 95% confidence interval
lower_bound, upper_bound = confidence_interval(sample_data, confidence_level=0.95)

# Output the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#Que.1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Function to perform Z-test
def z_test(sample, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate sample mean
    sample_mean = np.mean(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Interpret the result
    if p_value < alpha:
        result = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        result = "Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean."

    return z_statistic, p_value, result

# Generate a sample from a normal distribution
population_mean = 50
population_std = 10
sample_size = 30
sample = np.random.normal(loc=50, scale=10, size=sample_size)

# Perform Z-test
z_statistic, p_value, result = z_test(sample, population_mean, population_std, sample_size)

# Output the results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
#Que.4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to perform two-tailed Z-test and visualize the decision region
def z_test_and_visualize(sample, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate sample mean
    sample_mean = np.mean(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Determine critical Z-values for the decision region (two-tailed)
    critical_value = norm.ppf(1 - alpha / 2)  # Two-tailed critical Z value

    # Create the x-axis for the standard normal distribution plot
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)  # Probability density function for a standard normal distribution

    # Plot the standard normal distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label="Standard Normal Distribution")

    # Fill the critical regions (rejection regions) for two-tailed test
    plt.fill_between(x, 0, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label="Rejection Region")

    # Highlight the calculated Z-statistic
    plt.axvline(x=z_statistic, color='blue', linestyle='--', label=f'Z-statistic = {z_statistic:.2f}')

    # Mark critical values on the plot
    plt.axvline(x=-critical_value, color='green', linestyle='--', label=f'Critical Value = ±{critical_value:.2f}')
    plt.axvline(x=critical_value, color='green', linestyle='--')

    # Labels and title
    plt.title('Two-Tailed Z-test: Decision Region and Z-statistic')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Interpretation
    if p_value < alpha:
        result = f"Reject the null hypothesis: The sample mean is significantly different from the population mean (p-value = {p_value:.4f})."
    else:
        result = f"Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean (p-value = {p_value:.4f})."

    return z_statistic, p_value, result

# Simulate random data from a normal distribution
population_mean = 50  # Assume the true population mean is 50
population_std = 10   # Assume the true population standard deviation is 10
sample_size = 30      # Sample size
sample = np.random.normal(loc=50, scale=10, size=sample_size)  # Simulated sample data

# Perform the two-tailed Z-test and visualize
z_statistic, p_value, result = z_test_and_visualize(sample, population_mean, population_std, sample_size)

# Output the results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
#Que.5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to visualize and calculate Type 1 and Type 2 errors during hypothesis testing
def visualize_type_errors(sample_size, population_mean, population_std, alpha=0.05, true_population_mean=55):
    # Define critical value for a two-tailed test
    critical_value = norm.ppf(1 - alpha / 2)

    # Simulate the null hypothesis distribution (assuming H0 is true)
    null_mean = population_mean
    null_std = population_std / np.sqrt(sample_size)

    # Simulate the alternative hypothesis distribution (assuming H1 is true)
    alternative_mean = true_population_mean
    alternative_std = population_std / np.sqrt(sample_size)

    # Create x-axis for plotting the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y_null = norm.pdf(x, 0, 1)  # Probability density function for null distribution
    y_alternative = norm.pdf(x, (true_population_mean - population_mean) / population_std, 1)  # Alternative distribution

    # Plot the standard normal distribution for the null hypothesis
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Hypothesis Distribution (H0)", color="blue")
    plt.plot(x, y_alternative, label="Alternative Hypothesis Distribution (H1)", color="orange")

    # Shade the rejection region for Type 1 error (H0 is true but we reject it)
    plt.fill_between(x, 0, y_null, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label="Rejection Region (Type 1 Error)")

    # Shade the non-rejection region for Type 2 error (H1 is true but we fail to reject H0)
    plt.fill_between(x, 0, y_alternative, where=(x >= -critical_value) & (x <= critical_value), color='green', alpha=0.5, label="Non-Rejection Region (Type 2 Error)")

    # Add critical values lines
    plt.axvline(x=-critical_value, color='green', linestyle='--', label=f'Critical Value = ±{critical_value:.2f}')
    plt.axvline(x=critical_value, color='green', linestyle='--')

    # Add title and labels
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Type 1 Error calculation (false positive)
    type_1_error = 2 * (1 - norm.cdf(critical_value))

    # Type 2 Error calculation (false negative) under H1
    type_2_error = norm.cdf(critical_value - (true_population_mean - population_mean) / population_std) - norm.cdf(-critical_value - (true_population_mean - population_mean) / population_std)

    return type_1_error, type_2_error

# Set parameters for hypothesis testing
sample_size = 30
population_mean = 50    # The assumed population mean (under H0)
population_std = 10     # The population standard deviation
true_population_mean = 55  # The true mean (under H1)

# Call the function to visualize errors and get the error rates
type_1_error, type_2_error = visualize_type_errors(sample_size, population_mean, population_std, alpha=0.05, true_population_mean=true_population_mean)

# Output the calculated errors
print(f"Type 1 Error Rate (False Positive): {type_1_error:.4f}")
print(f"Type 2 Error Rate (False Negative): {type_2_error:.4f}")
```

```python
#Que.7 Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to perform paired sample T-test and visualize the comparison
def paired_t_test_and_visualize(before, after, alpha=0.05):
    # Perform the paired sample T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Output the T-statistic and p-value
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference between the two samples."
    else:
        result = "Fail to reject the null hypothesis: There is no significant difference between the two samples."

    # Plotting the before and after comparison
    plt.figure(figsize=(8, 6))
    x = np.arange(len(before))
    plt.plot(x, before, label="Before", marker='o', color='blue')
    plt.plot(x, after, label="After", marker='o', color='red')

    # Adding labels and title
    plt.title('Paired Sample Comparison (Before vs After)')
    plt.xlabel('Pair Index')
    plt.ylabel('Measurement')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_stat, p_value, result

# Simulate data: Before and After intervention
np.random.seed(0)
before = np.random.normal(loc=50, scale=10, size=30)  # Before intervention
after = before + np.random.normal(loc=5, scale=5, size=30)  # After intervention (with improvement)

# Perform the paired sample T-test and visualize the result
t_stat, p_value, result = paired_t_test_and_visualize(before, after, alpha=0.05)

# Output the interpretation
print(result)
```

```python
#Que.8 Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy import stats

# Function to perform Z-test and T-test, and compare the results
def z_test(sample1, sample2, pop_mean, pop_std, alpha=0.05):
    # Z-test assumes population parameters (mean, std) are known
    z_stat = (np.mean(sample1) - np.mean(sample2)) / (pop_std / np.sqrt(len(sample1)))
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed Z-test

    # Perform T-test
    t_stat, p_value_t = stats.ttest_ind(sample1, sample2)

    # Interpret the results
    result_z = "Reject the null hypothesis (Z-test): There is a significant difference." if p_value_z < alpha else "Fail to reject the null hypothesis (Z-test): No significant difference."
    result_t = "Reject the null hypothesis (T-test): There is a significant difference." if p_value_t < alpha else "Fail to reject the null hypothesis (T-test): No significant difference."

    return z_stat, p_value_z, t_stat, p_value_t, result_z, result_t

# Simulate two independent samples
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Sample 1: mean = 50, std = 10
sample2 = np.random.normal(loc=55, scale=10, size=30)  # Sample 2: mean = 55, std = 10

# Known population parameters for Z-test (mean and std)
pop_mean = 50  # Population mean (null hypothesis)
pop_std = 10   # Population standard deviation (known)

# Perform Z-test and T-test and compare the results
z_stat, p_value_z, t_stat, p_value_t, result_z, result_t = z_test(sample1, sample2, pop_mean, pop_std, alpha=0.05)

# Output the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (Z-test): {p_value_z:.4f}")
print(result_z)

print("\n")

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value (T-test): {p_value_t:.4f}")
print(result_t)
```

```python
#Que.9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence_level=0.95):
    # Sample mean and standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    n = len(data)  # Sample size

    # Degrees of freedom
    df = n - 1

    # Calculate the critical value for the confidence level
    if n > 30:  # Use Z-distribution for large samples
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = critical_value * (sample_std / np.sqrt(n))
    else:  # Use t-distribution for small samples
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df)
        margin_of_error = critical_value * (sample_std / np.sqrt(n))

    # Confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper

# Example usage: Simulate a dataset and calculate the confidence interval for the sample mean
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=30)  # Simulated data (mean=50, std=10, size=30)

# Calculate the 95% confidence interval for the sample mean
ci_lower, ci_upper = confidence_interval(data, confidence_level=0.95)

# Output the results
print(f"95% Confidence Interval for the Sample Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Que.10 Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def margin_of_error(data, confidence_level=0.95):
    # Calculate sample mean, standard deviation, and sample size
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)  # Sample size

    # Degrees of freedom
    df = n - 1

    # Determine the critical value (Z or t)
    if n > 30:  # Use Z-distribution for large samples
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    else:  # Use t-distribution for small samples
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(n))

    return margin_of_error

# Example usage: Simulate data and calculate margin of error for 95% confidence level
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=30)  # Simulated data (mean=50, std=10, size=30)

# Calculate margin of error
moe = margin_of_error(data, confidence_level=0.95)

# Output the result
print(f"Margin of Error for 95% Confidence Level: {moe:.2f}")
```

```python
#Que.11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(Disease) = prior probability of having the disease
    - sensitivity: P(Positive | Disease) = true positive rate
    - specificity: P(Negative | No Disease) = true negative rate

    Returns:
    - posterior: P(Disease | Positive)
    """
    p_disease = prior
    p_no_disease = 1 - prior

    # Likelihood of positive test
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity

    # Total probability of positive test
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Apply Bayes' Theorem
    posterior = (p_pos_given_disease * p_disease) / p_positive
    return posterior

# Parameters
prior = 0.01           # 1% prior probability (prevalence)
sensitivity = 0.99     # 99% true positive rate
specificity = 0.95     # 95% true negative rate

# Calculate posterior
posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of having the disease given a positive test: {posterior:.2%}")
```

```python
#Que.12 Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create the contingency table
data = np.array([[20, 15],   # Males: Coffee, Tea
                 [30, 25]])  # Females: Coffee, Tea

# Optional: Create a DataFrame for readability
df = pd.DataFrame(data, columns=['Coffee', 'Tea'], index=['Male', 'Female'])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(df)

# Output the results
print("Chi-square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("\nExpected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\n❌ Reject the null hypothesis: There is a significant association between gender and drink preference.")
else:
    print("\n✅ Fail to reject the null hypothesis: No significant association between gender and drink preference.")
```

```python
#Que.13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np
import pandas as pd

# Step 1: Create the observed data (contingency table)
# Example: Gender vs Drink Preference
observed = np.array([[20, 15],   # Males: Coffee, Tea
                     [30, 25]])  # Females: Coffee, Tea

# Convert to DataFrame for clarity
df_observed = pd.DataFrame(observed, columns=['Coffee', 'Tea'], index=['Male', 'Female'])

# Step 2: Calculate row totals, column totals, and grand total
row_totals = df_observed.sum(axis=1).values.reshape(-1, 1)  # shape: (2,1)
col_totals = df_observed.sum(axis=0).values.reshape(1, -1)  # shape: (1,2)
grand_total = df_observed.values.sum()

# Step 3: Calculate expected frequencies
expected = (row_totals @ col_totals) / grand_total  # Outer product

# Convert expected frequencies to DataFrame
df_expected = pd.DataFrame(expected, columns=['Coffee', 'Tea'], index=['Male', 'Female'])

# Step 4: Display results
print("Observed Frequencies:\n", df_observed)
print("\nExpected Frequencies (assuming independence):\n", df_expected.round(2))
```

```python
#Que.14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

from scipy.stats import chisquare

# Step 1: Define observed frequencies (e.g., results from a die roll)
observed = [8, 10, 9, 11, 12, 10]

# Step 2: Define expected frequencies (for a fair die, each outcome expected 10 times)
expected = [10] * 6

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Chi-square Statistic:", round(chi2_stat, 3))
print("P-value:", round(p_value, 4))

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("\n❌ Reject the null hypothesis: Observed data significantly differs from the expected distribution.")
else:
    print("\n✅ Fail to reject the null hypothesis: No significant difference between observed and expected.")
```
